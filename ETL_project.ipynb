```python
#dependencies
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

# Extract CSVs into DataFrames


```python
disneyplus_file = "Resources/disney_plus_shows.csv"
disneyplus_df = pd.read_csv(disneyplus_file, dtype = {"imdb_votes": str})
disneyplus_df.head()
```

```python
imdb_file = "Resources/IMDb ratings.csv"
imdb_df = pd.read_csv(imdb_file)
imdb_df.head()
```

# Transform Disney DataFrame


```python
#drop series so it is only movies
disneyplus_df = disneyplus_df.dropna(how = 'all') 

disneyplus_df.drop(disneyplus_df[disneyplus_df['type'] == "series"].index, inplace = True) 

disneyplus_df
```

```python
# Create a filtered dataframe from specific columns
disneyplus_cols = ["imdb_id", "title","imdb_rating", "imdb_votes"]
disneyplus_transformed= disneyplus_df[disneyplus_cols].copy()

# Clean the data by dropping duplicates and setting the index
disneyplus_transformed.drop_duplicates("imdb_id", inplace=True)
disneyplus_transformed.set_index("imdb_id", inplace=True)

disneyplus_transformed.head()
```

```python
disneyplus_transformed['imdb_votes'].dtypes
```

```python
disneyplus_transformed["imdb_votes"]=disneyplus_transformed["imdb_votes"].replace(",", "", regex=True)
```

```python
disneyplus_transformed["imdb_votes"]=disneyplus_transformed["imdb_votes"].astype(float)
```

```python
disneyplus_transformed = disneyplus_transformed.dropna(how='all')
disneyplus_transformed
```

# Transform IMDB DataFrame



```python
# Create a filtered dataframe from specific columns
imdb_cols = ["imdb_title_id", "allgenders_18age_avg_vote","allgenders_30age_avg_vote", "allgenders_45age_avg_vote","us_voters_rating","non_us_voters_rating"]
imdb_transformed= imdb_df[imdb_cols].copy()

# Rename the column headers
imdb_transformed = imdb_transformed.rename(columns={"imdb_title_id": "imdb_id",
                                                         "allgenders_18age_avg_vote": "eighteen_avg_vote",
                                                         "allgenders_30age_avg_vote": "thirty_avg_vote",
                                                         "allgenders_45age_avg_vote": "fourtyfive_avg_vote"})

# Clean the data by dropping duplicates and setting the index
imdb_transformed.drop_duplicates("imdb_id", inplace=True)
imdb_transformed.set_index("imdb_id", inplace=True)

imdb_transformed.head()
```

```python
connection_string = "postgres:@localhost:5432/Disney_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
disneyplus_transformed.to_sql(name='disneyplus', con=engine, if_exists='append', index=True)
```

```python
imdb_transformed.to_sql(name='imdb', con=engine, if_exists='append', index=True)
```
